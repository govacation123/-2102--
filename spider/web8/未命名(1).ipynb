```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = 1
url = [f'https://www.maotorino.it/en/collections/online-catalogue/?pag={page}&src&portal=mao&catalog_collection=131#filters-1']
# url = [f'https://www.maotorino.it/en/collections/online-catalogue/?pag={page}#filters-1']
data = []
```

```python
while url:
    print(f"正在爬取第{page}页")
    page += 1
    response = requests.get(url[0])
    soup = BeautifulSoup(response.content, 'html.parser')
    for item in soup.find_all('div', class_='grid-opera'):
        # print(f"正在处理第{i}条数据")
        # i += 1
        img_tag = item.find('img', class_='lazy')
        href = img_tag.get('data-src')
        des_tag = item.find('a')
        des_href = des_tag.get('href')
        name = item.find('h3').text
        response1 = requests.get(des_href)
        soup1 = BeautifulSoup(response1.content, 'html.parser')
        div = soup1.find('div', class_='row opera-details')
        la = div.find_all('label')
        details = div.find_all('div',class_='col-md-4')
        description = div.find('p').text if div.find('p') else 'No description'
        one_data = {'name':name,'img_href':href,'description':description,'desc_href':des_href}
        for detail in details[:-1]:
            inf = detail.text.split()
            label = inf[0]
            information = ''
            for inf_new in inf[1:]:
                information+=inf_new+' '
            information = information.strip()
            one_data[label]=information
        # print(one_data)
        data.append(one_data)
        # print(f'label: {label}')
        # print(f'href:{href}')

        # print(f'description:{description}')
        # print(f'Description: {description}')
    url.pop(0)
    pages = soup.find_all(attrs={'class': 'page-link'})
    next_page_href = pages[-1].get('href')
    next_page_text = pages[-1].text
    print(next_page_text)
    if next_page_text != "10":
        url.append(f'https://www.maotorino.it/en/collections/online-catalogue/{next_page_href}')
    else:
        print("爬完了")
#
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('china_collection.csv')
```
